# 1) Funções

A intuição sobre funções já é  bem familiar.

Uma **função** é um objeto utilizado para **fazer determinadas ações.**

Podemos ver uma função como, uma "caixinha" que pega uma entrada**(argumento)** , faz algum **processamento** e, então **retorna uma saída** (output).

Mas atenção! Não é obrigatório uma função ter  um argumento de entrada e uma saída! Um exemplo disso, é a função print(), ela não retorna uma variável.



Aprenderemos agora, como criar funções em python!

A estrutura de **definição de uma função** é:

    def nome_da_função(argumentos):
        instruções
      return saída    

Há 5 elementos fundamentais para criação de novas funções em python:



*   Primeiramente, usar o **def** para deixar claro que estamos **definindo** uma função.

*   Depois, damos um **nome** para nossa função.

*   Dentro do parênteses, falamos quais serão os **argumentos** da função.

*   Depois, explicitamos, qual é o **processamento** (instruções) que é executado pela função.

*  No fim, dizemos o que essa função vai **retornar**.


Sempre que quisermos usar uma função, basta **chama-lá**, dando os argumentos necessários para sua execução.

    nome_da_funao(argumentos)

**3. Uma função com dois argumentos mas sem retorno!**




```python
def cumprimenta_parte_do_dia(nome, parte_dia):

  if parte_dia == 'manhã':
    cumprimento = 'Bom dia'
  elif parte_dia == 'tarde':
    cumprimento = 'Boa tarde'
  elif parte_dia == 'noite':
    cumprimento = 'Boa noite'
  else:
    raise ValueError(f"A parte do dia informada ({parte_dia}) não é valida!")

  cumprimento = f"{cumprimento}, {nome}"
  print(cumprimento)
```

```python
cumprimenta_parte_do_dia('Ana', 'manhã')
```

Note, que ao passar "parte_dia" errado, o erro retornado é o criado em nossa função!

Além disso, os parâmetros devem ser passados na ordem correta que são pedidos na função!


**Mas e return?**

Todas as funções acima de fato fazem alguma operação, mas não conseguimos acesso ao resultado dessa operação! Veja um exemplo mais claro, uma função que calcula a soma de 2 números.

```python
def calcula_soma(n1, n2):
  soma = n1 + n2
  return soma
```

```python
calcula_soma(2,3)
```

Note que, a função faz os cálculos dos dois números, mas apenas exibe este resultado como um print!

**A variável soma existe apenas dentro da função!**

**Apenas o valor da variável é retornado, não o nome dela! Fora da função o nome da variável soma continua não existindo!**

Daí, basta retornar o valor da função em uma variável, por exemplo:
         
         resultado = calcula_soma(n1, n2)

```python
resultado1 = calcula_soma(2,5)
resultado2 = calcula_soma(10,1)
x = resultado1*resultado2
print(x)
```

Fazendo isso, conseguimos "usar" o return da função para fazer outras operações!


Vamos resolver nosso problema gerador! Agora vamos desenvolver uma calculadora, usando diversas funções. Assim, ficará claro, que uma função pode ser chamada dentro de outro função. Mostrando, assim, as boas práticas de programação, que são a simplicidade e a modulação de um algoritmo.

Vamos implementar função por função!

```python
def calcula_soma(n1, n2):
  soma = n1 + n2
  return soma
```

```python
def calcula_subtracao(n1,n2):
  subtracao = n1 -n2
  return subtracao
```

```python
def calcula_multiplicacao(n1,n2):
  multiplicacao = n1*n2
  return multiplicacao
```

```python
def calcula_divisao(n1,n2):
  divisao = n1/n2
  return divisao
```

```python
def calculadora(n1, n2, op):
  '''
    - n1 (float) -> numero 1
    - n2 (float) -> numero 2
    - op (str) -> operação a ser feita pela calculadora (soma, subtração, divisão e multiplicação)
  '''

  if op =='soma':
    return calcula_soma(n1,n2)

  elif op == 'subtracao':
    return calcula_subtracao(n1,n2)

  elif op == 'multiplicacao':
    return calcula_multiplicacao(n1,n2)
  elif op == 'divisao':
    return calcula_divisao(n1,n2)
  else:
    raise ValueError((f"Operação ({op}) não é suportada!"))

```

```python
num1 = float(input('Digite o primeiro número:'))
num2 = float(input('Digite o segundo número:'))
operacao = input('Digite a operação (soma, subtracao, divisao ou multiplicacao):')

x = calculadora(num1, num2, operacao)
print(x)
```

Função que calcula o fatorial de um número:

```python
def funcao(n):
    ans = 1
    for i in range(1, n+1):
        ans *= i
    return ans
```
